```python
# =============================================
# EDA INSTAGRAM MULTI-SEED - TFG
# =============================================
# Análisis Exploratorio de Datos (EDA) adaptado
# a una comunidad formada por varios nodos semilla
# =============================================

# =============================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# =============================================

import json
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import networkx as nx
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


plt.style.use("default")


# =============================================
# 2. CARGA DE LOS DATASETS (MULTI-SEED)
# =============================================

DATASETS = {
    "book": "nodo_book_593.json",
    "pique": "nodo_pique_595.json",
    "mercxdonx": "nodo_mercxdonx_523.json"
}

all_accounts = []

for seed_name, path in DATASETS.items():
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    for acc in data:
        acc["seed_origin"] = seed_name
        acc["is_seed"] = False
        all_accounts.append(acc)

print("===========================================")
print("CARGA DEL DATASET MULTI-SEED")
print("===========================================")
print(f"✔ Total de registros cargados (con duplicados): {len(all_accounts)}\n")

# =============================================
# 3. UNIFICACIÓN Y DEDUPLICACIÓN DE CUENTAS
# =============================================

df_raw = pd.DataFrame(all_accounts)

# Agrupamos por username (misma cuenta en distintos seeds)
df = (
    df_raw
    .groupby("username", as_index=False)
    .agg({
        "followers": "max",
        "following": "max",
        "posts_count": "max",
        "bio": "first",
        "is_private": "first",
        "is_verified": "first",
        "seed_origin": lambda x: list(set(x)),
        "followers_list": lambda x: sum([lst for lst in x if isinstance(lst, list)], []),
        "posts": lambda x: sum([lst for lst in x if isinstance(lst, list)], [])
    })
)

df["n_seeds"] = df["seed_origin"].apply(len)

print(f"✔ Número de cuentas únicas tras deduplicación: {len(df)}\n")

# =============================================
# 4. AJUSTE MENOR — COLUMNAS VISIBLES PARA EDA
# =============================================

if "followers_list" in df.columns:
    df_for_eda = df.drop(columns=["followers_list"])
else:
    df_for_eda = df.copy()

# =============================================
# 5. EDA DE VARIABLES NUMÉRICAS
# =============================================

def eda_numeric_features(df):
    numeric_cols = ["followers", "following", "posts_count"]

    print("\n===========================================")
    print("EDA DE VARIABLES NUMÉRICAS")
    print("===========================================\n")

    print("Estadísticas descriptivas:")
    display(df[numeric_cols].describe())

    for col in numeric_cols:
        plt.figure()
        plt.hist(df[col], bins=20)
        plt.xlabel(col)
        plt.ylabel("Número de cuentas")
        plt.title(f"Distribución de {col}")
        plt.show()

eda_numeric_features(df_for_eda)

# =============================================
# 6. FEATURE ENGINEERING: RATIOS
# =============================================

def add_ratio_features(df):
    df["follow_ratio"] = df["followers"] / (df["following"] + 1)
    return df

def analyze_follow_ratio(df):
    ratio = df["follow_ratio"]

    print("\n===========================================")
    print("ANÁLISIS DEL RATIO FOLLOWERS / FOLLOWING")
    print("===========================================\n")

    print(f"- Media: {ratio.mean():.2f}")
    print(f"- Mediana: {ratio.median():.2f}")
    print(f"- Mínimo: {ratio.min():.2f}")
    print(f"- Máximo: {ratio.max():.2f}\n")

    suspicious = df[df["follow_ratio"] < 0.3]
    print(f"✔ Cuentas con ratio < 0.3: {len(suspicious)}\n")

    plt.figure()
    plt.hist(ratio, bins=20)
    plt.xlabel("Followers / Following")
    plt.ylabel("Número de cuentas")
    plt.title("Distribución del ratio followers/following")
    plt.show()

df = add_ratio_features(df)
analyze_follow_ratio(df)

# =============================================
# 7. EXTRACCIÓN DE TEXTO (BIOS Y COMENTARIOS)
# =============================================

bios = []
comments = []

for acc in all_accounts:
    username = acc["username"]

    # BIO
    bio = acc.get("bio", "")
    if isinstance(bio, str) and bio.strip():
        bios.append({
            "username": username,
            "seed_origin": acc["seed_origin"],
            "bio": bio.strip()
        })

    # COMENTARIOS
    for post in acc.get("posts", []):
        for c in post.get("comments", []):
            text = c.get("text", "")
            if isinstance(text, str) and text.strip():
                comments.append({
                    "author": c.get("author"),
                    "target_account": username,
                    "seed_origin": acc["seed_origin"],
                    "text": text.strip()
                })

bios_df = pd.DataFrame(bios)
comments_df = pd.DataFrame(comments)

print("\n===========================================")
print("EXTRACCIÓN DE TEXTO")
print("===========================================\n")
print(f"- Bios extraídas: {len(bios_df)}")
print(f"- Comentarios extraídos: {len(comments_df)}")



# =============================================
# 7.1. CÁLCULO DE EMBEDDINGS SEMÁNTICOS
# =============================================

print("\n===========================================")
print("CÁLCULO DE EMBEDDINGS SEMÁNTICOS")
print("===========================================\n")

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Embeddings de bios
bio_embeddings = model.encode(
    bios_df["bio"].tolist(),
    show_progress_bar=True
)
bios_df["embedding"] = list(bio_embeddings)

# Embeddings de comentarios
comment_embeddings = model.encode(
    comments_df["text"].tolist(),
    show_progress_bar=True
)
comments_df["embedding"] = list(comment_embeddings)

print(f"- Dimensión del embedding: {bio_embeddings.shape[1]}")
print("✔ Embeddings generados correctamente\n")

# =============================================
# 7.2. SIMILITUD SEMÁNTICA GLOBAL DE COMENTARIOS
# =============================================

print("\n===========================================")
print("SIMILITUD SEMÁNTICA DE COMENTARIOS")
print("===========================================\n")

sample_size = min(500, len(comments_df))
sample_embeddings = np.vstack(
    comments_df.sample(sample_size, random_state=42)["embedding"].values
)

similarity_matrix = cosine_similarity(sample_embeddings)

upper_sim = similarity_matrix[np.triu_indices_from(similarity_matrix, k=1)]

print(f"- Similitud media entre comentarios: {upper_sim.mean():.4f}")
print(f"- Similitud máxima observada: {upper_sim.max():.4f}")
print(f"- Similitud mínima observada: {upper_sim.min():.4f}\n")

print(
    "Interpretación:\n"
    "- Valores altos indican comentarios muy similares entre sí.\n"
    "- Una similitud elevada y repetida es típica de automatización.\n"
)


# =============================================
# 7.3. DIVERSIDAD SEMÁNTICA POR CUENTA
# =============================================

def semantic_diversity(embeddings):
    if len(embeddings) < 2:
        return 0.0
    sim = cosine_similarity(embeddings)
    upper = sim[np.triu_indices_from(sim, k=1)]
    return 1 - upper.mean()

diversity_scores = []

for author, group in comments_df.groupby("author"):
    emb = np.vstack(group["embedding"].values)
    diversity_scores.append({
        "author": author,
        "semantic_diversity": semantic_diversity(emb),
        "num_comments": len(group)
    })

diversity_df = pd.DataFrame(diversity_scores)

print("\n===========================================")
print("DIVERSIDAD SEMÁNTICA POR CUENTA")
print("===========================================\n")
display(diversity_df.describe())


# =============================================
# 7.4. DISTRIBUCIÓN DE DIVERSIDAD SEMÁNTICA
# =============================================

plt.figure(figsize=(8, 4))
plt.hist(diversity_df["semantic_diversity"], bins=30)
plt.xlabel("Diversidad semántica")
plt.ylabel("Número de cuentas")
plt.title("Distribución de diversidad semántica de comentarios")
plt.show()

print(
    "\nInterpretación:\n"
    "- Valores cercanos a 0 indican lenguaje repetitivo.\n"
    "- Valores altos reflejan comportamiento humano más variado.\n"
)


# =============================================
# 8. FEATURE: PRESENCIA EN MÚLTIPLES SEEDS
# =============================================

print("\n===========================================")
print("PRESENCIA DE CUENTAS EN MÚLTIPLES NODOS SEMILLA")
print("===========================================\n")
display(df["n_seeds"].value_counts().sort_index())
print(
    "Interpretación:\n"
    "- Cuentas presentes en varios seeds actúan como puentes.\n"
    "- Estas cuentas son más relevantes estructuralmente.\n"
)

# =============================================
# 9. EDA DE LA RED DE FOLLOWERS (GLOBAL)
# =============================================

edges = []
valid_users = set(df["username"])

for _, row in df.iterrows():
    account_username = row.get("username")
    followers_data = row.get("followers_list", [])
    if isinstance(followers_data, list):
        for follower_username in followers_data:
            if isinstance(follower_username, str):
                edges.append((follower_username, account_username))

# Filtrar relaciones de usuarios inexistentes
edges = [(src, tgt) for src, tgt in edges if src in valid_users and tgt in valid_users]

print(f"\n✔ Número total de relaciones follower → cuenta: {len(edges)}")

def eda_network_basic(edges):
    if not edges:
        print("⚠ No se han detectado relaciones follower->cuenta.")
        return

    sources = [e[0] for e in edges]
    targets = [e[1] for e in edges]

    print("\n===========================================")
    print("EDA DE LA RED DE FOLLOWERS")
    print("===========================================\n")
    print(f"- Número total de relaciones (aristas): {len(edges)}")
    print(f"- Número de seguidores únicos: {len(set(sources))}")
    print(f"- Número de cuentas seguidas: {len(set(targets))}\n")

eda_network_basic(edges)

# =============================================
# 10. ANÁLISIS GLOBAL DEL GRAFO MULTI-SEED
# =============================================

G = nx.DiGraph()
G.add_edges_from(edges)

print("\n===========================================")
print("MÉTRICAS GLOBALES DE LA RED")
print("===========================================\n")
print(f"- Número de nodos: {G.number_of_nodes()}")
print(f"- Número de aristas: {G.number_of_edges()}")
print(f"- Densidad: {nx.density(G):.6f}")

clustering = nx.average_clustering(G.to_undirected())
print(f"- Clustering medio global: {clustering:.4f}")

print(
    "\nInterpretación:\n"
    "- Clustering bajo sugiere seguidores aislados.\n"
    "- Clustering moderado indica comunidad orgánica.\n"
)

# =============================================
# 11. ANÁLISIS POR SEED
# =============================================

for seed in df["seed_origin"].explode().unique():
    seed_df = df[df["seed_origin"].apply(lambda x: seed in x)]
    print(f"\n-------------------------------------------")
    print(f"ANÁLISIS DEL SEED: {seed}")
    print("-------------------------------------------")
    print(f"- Número de cuentas: {len(seed_df)}")
    print(f"- Número medio de followers: {seed_df['followers'].mean():.1f}")
    print(f"- Número medio de following: {seed_df['following'].mean():.1f}")
    print(f"- Número medio de posts: {seed_df['posts_count'].mean():.1f}")

# =============================================
# 12. ANÁLISIS DE VALORES FALTANTES
# =============================================

def eda_missing_values(df):
    """
    Analiza la presencia de valores nulos en el dataset.
    Permite justificar decisiones de imputación o limpieza.
    """

    print("\n===========================================")
    print("ANÁLISIS DE VALORES FALTANTES")
    print("===========================================\n")

    missing_ratio = df.isna().mean().sort_values(ascending=False)
    missing_ratio = missing_ratio[missing_ratio > 0]

    if missing_ratio.empty:
        print(
            "✔ No se detectan valores nulos relevantes en el dataset.\n"
            "Esto indica que el proceso de scraping ha sido consistente\n"
            "y no es necesaria una imputación sistemática.\n"
        )
    else:
        print(
            "⚠ Se detectan columnas con valores faltantes.\n"
            "Porcentaje de valores nulos por columna:\n"
        )
        display(missing_ratio.mul(100).round(2))


eda_missing_values(df)


# =============================================
# 13. ANÁLISIS DE OUTLIERS
# =============================================

def eda_outliers(df, column):
    """
    Visualiza outliers mediante boxplot.
    Especialmente relevante en redes sociales.
    """

    print("\n-------------------------------------------")
    print(f"OUTLIERS EN LA VARIABLE: {column}")
    print("-------------------------------------------")

    plt.figure()
    plt.boxplot(df[column].dropna(), vert=False)
    plt.xlabel(column)
    plt.title(f"Outliers en {column}")
    plt.show()

    print(
        "Interpretación:\n"
        "- Valores extremos pueden corresponder a:\n"
        "  • cuentas muy populares (influencers)\n"
        "  • cuentas automatizadas con following masivo\n"
        "- Por tanto, NO se eliminan automáticamente,\n"
        "  pero deben tenerse en cuenta en el modelado.\n"
    )


eda_outliers(df, "followers")
eda_outliers(df, "following")
eda_outliers(df, "posts_count")


# =============================================
# 14. ANÁLISIS DE CORRELACIÓN
# =============================================

def eda_correlation(df):
    """
    Analiza correlaciones entre variables numéricas.
    Útil para detectar redundancia de features.
    """

    print("\n===========================================")
    print("ANÁLISIS DE CORRELACIÓN ENTRE VARIABLES")
    print("===========================================\n")

    numeric_df = df.select_dtypes(include="number")

    if numeric_df.shape[1] < 2:
        print("No hay suficientes variables numéricas para correlación.")
        return

    corr = numeric_df.corr()

    plt.figure(figsize=(10, 8))
    plt.imshow(corr)
    plt.colorbar()
    plt.xticks(range(len(corr)), corr.columns, rotation=90)
    plt.yticks(range(len(corr)), corr.columns)
    plt.title("Matriz de correlación entre variables numéricas")
    plt.show()

    print(
        "Interpretación:\n"
        "- Correlaciones altas pueden indicar redundancia.\n"
        "- Este análisis ayuda a decidir si eliminar variables\n"
        "  o aplicar reducción de dimensionalidad en fases posteriores.\n"
    )


eda_correlation(df)


# =============================================
# 15. EDA DE LA RED DE FOLLOWERS
# =============================================

print("===========================================")
print("EXTRACCIÓN DE RELACIONES DE LA RED")
print("===========================================\n")

edges = []

for _, row in df.iterrows():
    account_username = row.get("username")
    followers_data = row.get("followers_list", [])

    # followers_list es una lista de strings
    if isinstance(followers_data, list):
        for follower_username in followers_data:
            if isinstance(follower_username, str):
                edges.append((follower_username, account_username))

print(f"✔ Número total de relaciones follower → cuenta extraídas: {len(edges)}")


def eda_network_basic(edges):
    """
    Análisis exploratorio básico de la red de followers
    antes de construir el grafo.
    """

    print("\n===========================================")
    print("EDA DE LA RED DE FOLLOWERS")
    print("===========================================\n")

    if not edges:
        print("⚠ No se han detectado relaciones follower->cuenta.")
        return

    sources = [e[0] for e in edges]
    targets = [e[1] for e in edges]

    print(f"- Número total de relaciones (aristas): {len(edges)}")
    print(f"- Número de seguidores únicos: {len(set(sources))}")
    print(f"- Número de cuentas seguidas: {len(set(targets))}\n")

    degree_in = Counter(targets)
    degrees = list(degree_in.values())

    print(
        "Distribución del grado de entrada (número de seguidores por cuenta):\n"
        "- El grado es una medida fundamental para análisis de comunidades.\n"
    )

    plt.figure()
    plt.hist(degrees, bins=30)
    plt.xlabel("Grado de entrada (followers)")
    plt.ylabel("Número de cuentas")
    plt.title("Distribución del grado de entrada")
    plt.show()

    print(
        "Interpretación:\n"
        "- Muchas cuentas con grado bajo indican usuarios poco relevantes.\n"
        "- Cuentas con grado muy alto pueden actuar como hubs.\n"
        "- Esta información se usará para justificar limpieza del grafo.\n"
    )


eda_network_basic(edges)


```

```python

# =============================================
# 16. ANÁLISIS DE COMUNIDADES (MULTI-SEED) + VISUALIZACIÓN
# =============================================

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
import numpy as np

# Importar best_partition desde el submódulo community_louvain
import community.community_louvain as cl

# -----------------------------
# Convertimos grafo dirigido a no dirigido
# -----------------------------
G_undirected = G.to_undirected()

# -----------------------------
# Aplicamos el algoritmo Louvain
# -----------------------------
partition = cl.best_partition(G_undirected)

# Asignamos comunidad a cada nodo
community_df = pd.DataFrame.from_dict(partition, orient='index', columns=['community']).reset_index()
community_df.rename(columns={'index':'username'}, inplace=True);

# Merge con df principal
df_community = df.merge(community_df, on='username', how='left')

# Mostrar resumen de comunidades
community_counts = df_community['community'].value_counts().sort_index()
print("\n===========================================")
print("COMUNIDADES DETECTADAS")
print("===========================================\n")
print(f"- Número de comunidades detectadas: {df_community['community'].nunique()}")
display(community_counts)

# -----------------------------
# 16.1. Análisis de características por comunidad
# -----------------------------
for com_id, com_group in df_community.groupby("community"):
    print(f"\n-------------------------------------------")
    print(f"COMUNIDAD {com_id}")
    print("-------------------------------------------")
    print(f"- Número de cuentas: {len(com_group)}")
    print(f"- Número medio de followers: {com_group['followers'].mean():.1f}")
    print(f"- Número medio de following: {com_group['following'].mean():.1f}")
    print(f"- Número medio de posts: {com_group['posts_count'].mean():.1f}")
    print(f"- Número medio de seeds por cuenta: {com_group['n_seeds'].mean():.1f}")
    print(f"- Usuarios con follow_ratio < 0.3: {(com_group['follow_ratio'] < 0.3).sum()}")

    # Diversidad semántica media
    if 'embedding' in bios_df.columns:
        users_in_com = com_group['username'].tolist()
        bios_embeddings = [vec for u, vec in zip(bios_df['username'], bios_df['embedding']) if u in users_in_com]
        if len(bios_embeddings) > 1:
            bios_embeddings = np.vstack(bios_embeddings)
            diversity = 1 - cosine_similarity(bios_embeddings).mean()
            print(f"- Diversidad semántica media de bios: {diversity:.3f}")

# -----------------------------
# 16.2. Identificar nodos centrales (hubs) por comunidad
# -----------------------------
community_hubs = {}
for com_id, com_group in df_community.groupby('community'):
    nodes_in_com = com_group['username'].tolist()
    subgraph = G_undirected.subgraph(nodes_in_com)
    if len(subgraph) > 0:
        hub_node = max(subgraph.degree, key=lambda x: x[1])[0]  # nodo con mayor grado
        community_hubs[com_id] = hub_node

print("\nNODOS CENTRALES (HUBS) POR COMUNIDAD:")
for com_id, hub in community_hubs.items():
    print(f"- Comunidad {com_id}: {hub}")

# =============================================
# 16.3. Visualización del grafo por comunidades
# =============================================

colors = plt.colormaps["tab20"]  # colores para comunidades

# Layout jerárquico por comunidad para evitar solapamiento
pos = {}
x_offset = 0
y_spacing = 0
for comm in sorted(community_counts.index):
    nodes = [n for n, c in partition.items() if c == comm]
    comm_pos = nx.spring_layout(G.subgraph(nodes), seed=42)
    for n, (x, y) in comm_pos.items():
        pos[n] = (x + x_offset, y + y_spacing)
    x_offset += 5  # separa comunidades horizontalmente

plt.figure(figsize=(18, 12))

# Dibujar nodos y edges
for comm in sorted(community_counts.index):
    nodes = [n for n, c in partition.items() if c == comm]
    degrees = np.array([G.degree(n) for n in nodes])
    max_deg = degrees.max() if degrees.size > 0 else 1

    # Diferenciar hubs de nodos regulares
    node_sizes = []
    node_colors = []
    for n, deg in zip(nodes, degrees):
        if n == community_hubs[comm]:
            node_sizes.append(800)      # hub más grande
            node_colors.append("red")   # rojo para nodo central
        else:
            node_sizes.append(50 + 300*(deg/max_deg))  # tamaño proporcional al grado
            node_colors.append(colors(comm))           # color por comunidad

    nx.draw_networkx_nodes(
        G, pos,
        nodelist=nodes,
        node_size=node_sizes,
        node_color=node_colors,
        alpha=0.8,
        label=f"Comunidad {comm}"
    )

# Dibujar aristas
nx.draw_networkx_edges(G, pos, alpha=0.2, width=0.5)

# Etiquetas solo de hubs
nx.draw_networkx_labels(G, pos, labels={n:n for n in community_hubs.values()},
                        font_size=12, font_color="black", font_weight="bold")

plt.title("Grafo de usuarios de Instagram por comunidades (solo hubs etiquetados)")
plt.axis("off")
plt.show()

# =============================================
# 16.4. ANÁLISIS DE COHESIÓN POR COMUNIDAD
# =============================================

print("\n===========================================")
print("ANÁLISIS DE COHESIÓN POR COMUNIDAD")
print("===========================================\n")

cohesion_results = []

for com_id, com_group in df_community.groupby("community"):
    nodes_in_com = com_group["username"].tolist()
    subgraph = G_undirected.subgraph(nodes_in_com)

    n_nodes = subgraph.number_of_nodes()
    n_edges = subgraph.number_of_edges()

    if n_nodes < 3:
        continue  # comunidades demasiado pequeñas no son informativas

    # 1. Densidad interna
    density = nx.density(subgraph)

    # 2. Clustering medio
    clustering = nx.average_clustering(subgraph)

    # 3. Centralización de grado (proxy)
    degrees = np.array([deg for _, deg in subgraph.degree()])
    max_degree = degrees.max()
    degree_centralization = max_degree / (n_nodes - 1)

    # 4. Proporción de nodos hoja (grado = 1)
    leaf_ratio = np.sum(degrees == 1) / n_nodes

    cohesion_results.append({
        "community": com_id,
        "n_nodes": n_nodes,
        "density": density,
        "avg_clustering": clustering,
        "degree_centralization": degree_centralization,
        "leaf_ratio": leaf_ratio
    })

    print(f"-------------------------------------------")
    print(f"COMUNIDAD {com_id}")
    print("-------------------------------------------")
    print(f"- Número de nodos: {n_nodes}")
    print(f"- Densidad interna: {density:.5f}")
    print(f"- Clustering medio: {clustering:.4f}")
    print(f"- Centralización de grado: {degree_centralization:.3f}")
    print(f"- Proporción de nodos hoja: {leaf_ratio:.2f}")

    print(
        "Interpretación:\n"
        "- Densidad baja + alta centralización → estructura hub-and-spoke.\n"
        "- Clustering bajo → seguidores aislados entre sí.\n"
        "- Muchos nodos hoja → posible red artificial.\n"
    )

cohesion_df = pd.DataFrame(cohesion_results)

print("\n===========================================")
print("RESUMEN GLOBAL DE COHESIÓN")
print("===========================================\n")
display(cohesion_df)
```

```python
# =============================================
# 28 INCORPORACIÓN GROUND TRUTH
# =============================================

print("\n" + "="*50)
print("GROUND TRUTH — FAKES CONFIRMADAS")
print("="*50)

fake_df = pd.read_json("nodo_fake.json")

ml_df["is_fake_real"] = ml_df["username"].isin(fake_df["username"]).astype(int)

total_accounts = len(ml_df)
total_fakes = ml_df["is_fake_real"].sum()
fake_ratio = ml_df["is_fake_real"].mean()

print(f"Total cuentas            : {total_accounts}")
print(f"Fakes confirmadas        : {total_fakes}")
print(f"Proporción real de fake  : {fake_ratio:.2%}")


# =============================================
# 29 EVALUACIÓN INDIVIDUAL DE MODELOS
# =============================================

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score
)

def evaluate_model(pred_column, model_name):

    y_true = ml_df["is_fake_real"]
    y_pred = ml_df[pred_column].astype(int)

    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print("\n" + "-"*50)
    print(f"MODELO: {model_name}")
    print("-"*50)

    print("Matriz de confusión")
    print(f"  TN: {tn:6}   FP: {fp:6}")
    print(f"  FN: {fn:6}   TP: {tp:6}")

    print("\nMétricas")
    print(f"  Precision : {precision:.4f}")
    print(f"  Recall    : {recall:.4f}")
    print(f"  F1-score  : {f1:.4f}")

    print("\nClassification report")
    print(classification_report(y_true, y_pred, digits=4))


evaluate_model("dbscan_anomaly", "DBSCAN")
evaluate_model("ocsvm_anomaly", "OneClassSVM")
evaluate_model("elliptic_anomaly", "EllipticEnvelope")
evaluate_model("lof_anomaly", "LOF")
evaluate_model("consensus_anomaly", "Consenso ≥2")


# =============================================
# 30 COMPARACIÓN GLOBAL DE MODELOS
# =============================================

print("\n" + "="*50)
print("COMPARACIÓN GLOBAL (ordenado por F1)")
print("="*50)

results = []

models = {
    "DBSCAN": "dbscan_anomaly",
    "OCSVM": "ocsvm_anomaly",
    "EllipticEnvelope": "elliptic_anomaly",
    "LOF": "lof_anomaly",
    "Consenso ≥2": "consensus_anomaly"
}

for name, col in models.items():

    y_true = ml_df["is_fake_real"]
    y_pred = ml_df[col].astype(int)

    results.append({
        "Modelo": name,
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1": f1_score(y_true, y_pred, zero_division=0)
    })

results_df = (
    pd.DataFrame(results)
    .sort_values("F1", ascending=False)
    .reset_index(drop=True)
)

display(results_df)

model_errors = []

models_all = {
    "DBSCAN": "dbscan_anomaly",
    "OCSVM": "ocsvm_anomaly",
    "Elliptic": "elliptic_anomaly",
    "LOF": "lof_anomaly",
    "Consenso": "consensus_anomaly"
}

for name, col in models_all.items():

    y_true = ml_df["is_fake_real"]
    y_pred = ml_df[col].astype(int)

    errors = (y_true != y_pred).sum()

    model_errors.append({
        "Modelo": name,
        "Errores": errors,
        "Error_rate": errors / len(y_true)
    })

errors_df = pd.DataFrame(model_errors)
errors_df.to_csv("model_errors.csv", index=False)

# =============================================
# 31 VISUALIZACIÓN COMPARATIVA
# =============================================

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.barplot(
    data=results_df.melt(id_vars="Modelo"),
    x="Modelo",
    y="value",
    hue="variable"
)

plt.title("Precision vs Recall vs F1")
plt.ylabel("Score")
plt.ylim(0,1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =============================================
# 32 MATRIZ DE CONFUSIÓN — MEJOR MODELO
# =============================================

best_model_name = results_df.loc[0, "Modelo"]
best_column = models[best_model_name]

y_true = ml_df["is_fake_real"]
y_pred = ml_df[best_column].astype(int)

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Pred Normal", "Pred Fake"],
    yticklabels=["Real Normal", "Real Fake"]
)

plt.title(f"Matriz de Confusión — {best_model_name}")
plt.ylabel("Real")
plt.xlabel("Predicción")
plt.tight_layout()
plt.show()
```
